```python
# Preparing train data
train_data = [
    ["Aragorn was the heir of Isildur", 1],
    ["Frodo was the heir of Isildur", 0],
]
train_df = pd.DataFrame(train_data)
train_df.columns = ["text", "labels"]

# Preparing eval data
eval_data = [
    ["Theoden was the king of Rohan", 1],
    ["Merry was the king of Rohan", 0],
]
eval_df = pd.DataFrame(eval_data)
eval_df.columns = ["text", "labels"]


# Create a ClassificationModel
```

```python
df = pd.read_csv('/content/hate.csv',encoding='ISO-8859-1')
```

```python
df.head()
```

```python
df.drop(columns={'Unnamed: 0'},inplace=True)
```

```python
df.head()
```

```python
df = df[['comment','label']]
```

```python
df['label'].value_counts()
```

```python
df = df[df['label'] != 'O']
```

```python
df['label'].value_counts()
```

```python
import re
```

```python
def change_label(label):
  if label == 'N':
    return 0
  elif label == 'P':
    return 1
```

```python
def preprocess_text(text):
    # Lowercase the text
    text = text.lower()

    # Remove special characters, punctuation, and numbers
    text = re.sub(r'[^\w\s]', '', text)

    return text
```

```python
df['label'] = df['label'].apply(change_label)
df['comment'] = df['comment'].apply(preprocess_text)
```

```python
df['label'].value_counts()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(df['comment'], df['label'], test_size=0.33, random_state=42)
```

```python
model_args = ClassificationArgs(num_train_epochs=1)
```

```python
model_args = {
    "output_dir": "new_output_directory/",  # Specify a new output directory
    # Other model configuration options
}
```

```python


model = ClassificationModel(
    "roberta", "roberta-base", args=model_args
)

# Train the model
model.train_model(train_df)
```

```python
# Evaluate the model
result, model_outputs, wrong_predictions = model.eval_model(test_df)
```
